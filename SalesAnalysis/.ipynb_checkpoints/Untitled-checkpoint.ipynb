```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.float_format', lambda x: '%.5f' % x)
from matplotlib.ticker import ScalarFormatter
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

### Merging 12 months of sales data to one csv

```python
All_records=pd.DataFrame()
for data in os.listdir("./Sales_Data"):
    record=pd.read_csv("./Sales_Data/"+data)
    All_records=pd.concat([All_records,record])
    
All_records.to_csv("All_months_data.csv", index=False)
```

```python
df=pd.read_csv("All_months_data.csv")
df
```

### What is the best month for sales and how much was earned that month

```python
df.isnull().sum()
```

```python
df=df[df['Order Date'].str[0:2]!='Or']
```

```python
df=df.dropna().reset_index()
```

```python
df['Month']=(df['Order Date'].str[0:2]).astype('int')
df['Date']=(df['Order Date'].str[3:5]).astype('int')
df['Hrs']=(df['Order Date'].str[9:11]).astype('int')
df
```

```python
df['Quantity Ordered']=df['Quantity Ordered'].astype('int')
df['Price Each']=df['Price Each'].astype('float')
df['Amount']=df['Quantity Ordered']*df['Price Each']
df
```

### Best Month for Sales

```python
MonthSales=df.groupby([('Month')])['Price Each','Amount'].sum().reset_index()
MonthSales
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=MonthSales, x='Month', y="Amount")
```

### what city had the highest number of sales

```python
def getCity(address):
    return address.split(',')[1]

def getState(address):
    return address.split(',')[2].split(" ")[1]

df['City']=df['Purchase Address'].apply(lambda x: getCity(x)+' '+ f"({getState(x)})")
df.head()
```

```python
CitySales=df.groupby([('City')])['Amount'].sum().reset_index()
CitySales
```

```python
plt.figure(figsize=(12,10))
plt.bar(CitySales['City'].unique(),CitySales['Amount'])
plt.xticks(CitySales['City'].unique(),rotation='vertical', size=12)
ax=plt.gca()
y_fmt = ScalarFormatter(useOffset=False)
y_fmt.set_scientific(False)
ax.yaxis.set_major_formatter(y_fmt)
# fmt=matplotlib.ticker.ScalarFormatter(useOffset=False)
# fmt.set_scientific(False)
# ax.yaxis.set_major_formatter(fmt)
#plt.ticklabel_format(style='plain')
plt.ylabel('Amount',size=12)
plt.xlabel('Cities', size=12)
```

### What time should we display the advert to maximize the likelihood of customer buying the product

```python
plt.figure(figsize=(12,8))
TimeData=df.groupby([('Hrs')])['Quantity Ordered'].count().reset_index()
sns.barplot(x='Hrs',y='Quantity Ordered', data=TimeData)
```

### What products are often sold together

```python
duplicates=df[df['Order ID'].duplicated(keep=False)]
duplicates
```

```python
duplicates['Grouped']=duplicates.groupby([('Order ID')])['Product'].transform(lambda x: ','.join(x))
duplicates
```

```python
## drop duplicates
duplicates=duplicates[['Order ID','Grouped']].drop_duplicates()
duplicates
```

```python
from itertools import combinations
from collections import Counter

count=Counter()

for row in duplicates['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2))) #put 3 to see most common in 3 items
    
for key,value in count.most_common(10):
    print(key,value)
```

### What product sold the most and why do you thing it sold the most

```python
Most_sold=df.groupby(['Product'])['Quantity Ordered'].sum().reset_index()
```

```python
plt.figure(figsize=(12,10))
plt.bar(Most_sold['Product'],Most_sold['Quantity Ordered'])
plt.xticks(Most_sold['Product'],rotation='vertical', size=12)
plt.ylabel('Quantity Ordered', size=16)
plt.xlabel('Product', size=16)
plt.show()
```

```python
prices=df.groupby(['Product']).mean()['Price Each']
prices
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))
ax2 = ax1.twinx()
ax1.bar(Most_sold['Product'],Most_sold['Quantity Ordered'],color='green')
ax2.plot(Most_sold['Product'],prices, 'b-')
ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price in (usd)', color='b')
ax1.set_xticklabels(Most_sold['Product'],rotation='vertical', size=10)
plt.show()
```
